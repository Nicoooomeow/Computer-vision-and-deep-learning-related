```python
# Lab1 Task4
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('Cat.png', 0)

blur_img = cv2.GaussianBlur(img, (3, 3), 1)
output = (img + 1.25 * (blur_img.astype('int8') - img.astype('int8'))).astype('uint8')

# Contrast stretch
a = 0.0
b = 255.0
c = np.min(output)
d = np.max(output)
img_0 = (output - c) * ((b - a) / (d - c)) + a

plt.figure()
plt.subplot(1, 2, 1)
plt.title('Original image'), plt.xticks([]), plt.yticks([])
plt.imshow(img, 'gray', vmin=0, vmax=255)
plt.subplot(1, 2, 2)
plt.title('Sharpen image'), plt.xticks([]), plt.yticks([])
plt.imshow(img_0, 'gray', vmin=0, vmax=255)
plt.show()
```
